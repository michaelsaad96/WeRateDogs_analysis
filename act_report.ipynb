# Playing with Dogs on Twitter

We wanted to find out which dog breeds get the most likes on WeRateDogs twitter account. We analyzed the tweets made by the account using the number of likes of each tweet along with the breed of the dog (which was extracted from the dog image using a neural network) and arrived at the following table.

| Breeds        | Likes  |   
| ------------- |--------|
| Saluki        | 24060  |
| French Bulldog| 17913  |
| Afghan Hound  | 17337.5|

This shows that the most favorited dog breed is the Saluki with 24060 favorites on average per appearance, followed by the French Bulldog with 17913 likes, and then the Afghan Hound in third place with 17337.5 likes on average per appearance.

It was interesting to find out which tweet length gets the most retweets. It shows whether the long or short text gets retweeted more often. We divided the tweet length of the tweets in our dataset into quartiles, and measured the number of retweets that each quartile received in total. This gave us the following table.

| Tweet Length| Retweet Count|
| ------------|--------------|
| 106 - 132   | 2460898      |
| 140 - 167   | 2281544      |
| 132 - 140   | 1857760      |
| 36 - 106    | 1709054      |

From the table we can see that tweets that are between 106 and 132 characters long receive the most retweets. Tweets that are between 140 and 167 characters come in at a close second place.

WeRateDogs uses a set of 'ranks' or 'stages' that distinguish between the different shapes and sizes of dogs. Those are 'puppo', 'floofer', 'pupper', and 'doggo'. Some dogs in our tweet dataset were assigned multiple ranks, while others were not assigned any rank. The table below shows how many times each dog rank appeared in our dataset, as well as the average rating it received.

| Dog Rank| Apearance Count| Average Rating (out of 10)|
| --------|----------------| -----------               |
| Puppo   | 23             | 12.04                     |
| Floofer | 7              | 11.85                     |
| None    | 1614           | 11.73                     |
| Doggo   | 64             | 11.73                     |
| Pupper  | 199            | 10.57                     |

It appears that puppo gets the highest rating, on average.
